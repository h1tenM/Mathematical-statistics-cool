The Distribution for the minimum of the sample taken from a population modeled with r.v.s Exponentially distributed (paramenter mean = B) is distributed Exponentially with mean = B/n, where n is the sample size. So cool!

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

population_size = 100000
sample_size = 20
num_samples = 5000
B = 2  # Mean of the original exponential distribution 
scale = B / sample_size  # Scale parameter for the minimum distribution


population = np.random.exponential(scale = B, size = population_size)

sample_mins = np.array([np.min(np.random.choice(population, sample_size, replace=False)) for _ in range(num_samples)])
print(sample_mins.shape)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
# density=True return probability density
plt.hist(population, bins=100, color='skyblue', density=True, edgecolor='black')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Probabiliy desity')

plt.subplot(1, 2, 2)
plt.hist(sample_mins, bins=100, color='lightgreen', edgecolor='black',density=True, label='Sample Minimum Distribution')
plt.xlabel('Minimum of Samples')
plt.ylabel('Frequency')

# Generate x values (domain for the PDF)
x = np.linspace(expon.ppf(0.01, scale=scale), expon.ppf(0.99, scale=scale), 1000) 
# ppf return the value for which propbaility is given as argument
plt.plot(x, expon.pdf(x, scale=scale), color='red', label='expected distribution of minimum of samples')
plt.title(f'Sampling Distribution of the Minimum\n(Sample Size = {sample_size}, Num Samples = {num_samples})')
plt.legend()

plt.tight_layout()
plt.show()


```
