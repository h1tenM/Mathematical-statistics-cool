```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon


true_mean = 26 # can be anything obviously
sample_size = 100
num_samples = 1000 
rate = 1 / true_mean  # rateparameter


samples = np.random.exponential(scale=true_mean, size=(num_samples, sample_size))
# print(sampple.shape)

sample_means = np.mean(samples, axis=1)

# Apply scaling to create pivotal statistic; Expo(1) as variables are iid, so sample mean is unbiased and large n
scaled_sample_means = sample_means * rate  # pivotal statistic = mean * rate(easy to normalize)

# confidence coefficient = 95%, alpha = 0.05
lower_bound = np.percentile(scaled_sample_means, 2.5)
upper_bound = np.percentile(scaled_sample_means, 97.5)

 
plt.figure(figsize=(12, 6))

# Plot original sample means
plt.subplot(1, 2, 1)
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.axvline(true_mean, color='r', linestyle='dashed', linewidth=2)
plt.title("Sample Means (Original Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

# Plot scaled sample means
plt.subplot(1, 2, 2)
plt.hist(scaled_sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.axvline(1, color='r', linestyle='dashed', linewidth=2)
plt.title("Scaled Sample Means (Pivotal Statistic)")
plt.xlabel("Scaled Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Output the confidence intervals for the pivotal method
print(f"95% Confidence Interval using the Pivotal Method: ({lower_bound:.3f}, {upper_bound:.3f})")
```
