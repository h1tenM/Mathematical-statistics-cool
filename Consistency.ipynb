This program illustrates the concept of consistency. In particular it illustrates that the sample mean is a consistent estimator of the population mean.

In this example we use the normal distribution as the distribution of the population.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the true population parameters
population_mean = 50
population_std = 10

# Generate a large population to simulate the true distribution
np.random.seed(42)  # For reproducibility
population = np.random.normal(loc=population_mean, scale=population_std, size=1000000)
#population = np.random.exponential(scale=population_mean, size=1000000)

# Function to estimate the mean with increasing sample sizes
def illustrate_consistency():
    sample_sizes = np.arange(10, 2001, 10)  # Sample sizes from 10 to 1000 in steps of 10
    estimated_means = []

    for n in sample_sizes:
        sample = np.random.choice(population, size=n, replace=False)  # Random sample from population
        estimated_means.append(np.mean(sample))

    # Plot the estimated means
    plt.figure(figsize=(10, 6))
    plt.plot(sample_sizes, estimated_means, label="Estimated Mean", marker='o', markersize=4, alpha=0.7)
    plt.axhline(y=population_mean, color='r', linestyle='--', label="True Mean")
    plt.title("Consistency of the Sample Mean")
    plt.xlabel("Sample Size")
    plt.ylabel("Estimated Mean")
    plt.legend()
    plt.grid()
    plt.show()

# Run the illustration
illustrate_consistency()
```
