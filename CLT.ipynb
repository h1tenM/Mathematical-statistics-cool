```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the parameters
population_size = 100000
sample_size = 20
num_samples = 5000

# Generate a non-normal population (e.g., exponential distribution)
population = np.random.exponential(scale=2, size=population_size)
# print(population.shape)

# Draw samples and compute their means
sample_means = [np.mean(np.random.choice(population, sample_size, replace=False)) for _ in range(num_samples)]

sample_means = np.array(sample_means)
# print(sample_means.shape)


# Plot the original population distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
# density=True return probability density
plt.hist(population, bins=100, color='skyblue', density=True, edgecolor='black')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')


# Plot the sampling distribution of the mean
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=100, color='lightgreen', edgecolor='black',density = True)
plt.title(f'Sampling Distribution of the Mean\n(Sample Size = {sample_size}, Num Samples = {num_samples})')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')


# Add the theoretical density line
mean = np.mean(population)
std_dev = np.std(population) / np.sqrt(sample_size)
x = np.linspace(min(sample_means), max(sample_means), 1000)
plt.plot(x, norm.pdf(x, loc=mean, scale=std_dev), color='red', label='Theoretical Density')
plt.title(f'Sampling Distribution of the Mean\n(Sample Size = {sample_size}, Num Samples = {num_samples})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

Here we draw samples from a large population. Note that the elements of the sample are not exactly independent because they are drawn from population without replacement. However, since the population size is large the elements of a sample are approximately independent.

We draw many samples of a given size and for every sample we calculate its mean. We do it many times and then we will plot the histogram of the mean to see the sample mean distribution.
